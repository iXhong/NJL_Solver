```python
import numpy as np
import matplotlib.pyplot as plt
```

## use Julia to calc and python to plot

data1 is M at  T=10 MeV

```python
data1 = np.load("T_10GeV_M_data.npz")
```

```python
# T_grid = data1["T_grid"]
mu_vals = data1["mu_vals"]
M_vals = data1["M_vals"]
```

```python
print(mu_vals.shape)
```

```python
fig,ax = plt.subplots()

ax.plot(mu_vals,M_vals,color='b')
# ax.legend()
ax.set_xlabel(r"\mu Chemical Potential")
ax.set_ylabel("M Effective Mass")
ax.set_title(r"M - \mu at T=10MeV")
ax.grid()
plt.show()
```

### 3d Plot

```python
# from matplotlib import cm
```

```python
# M data of T & mu
data = np.load("M_data.npz")
```

```python
T_grid = data["T_grid"]
mu_grid = data["mu_grid"]
M_grid = data["M_grid"]
```

```python
print(M_grid.shape)
```

```python
fig = plt.figure(figsize=(12,8))
# ax = plt.axes(projection='3d')
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface( mu_grid,T_grid, M_grid, cmap="rainbow")
ax.set_xlabel("mu Chemical Potential MeV")
ax.set_ylabel("T MeV")
ax.set_zlabel("M Effective Mass MeV")
ax.set_zlim(0,400)
ax.view_init(elev=15,azim=-25)
fig.colorbar(surf,shrink=0.5)
```
